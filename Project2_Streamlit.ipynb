```python

%%writefile /content/CustomerSegmentation/app.py

import streamlit as st
import pandas as pd
import numpy as np
import pickle

# SETTING PAGE CONFIG

st.set_page_config(
    page_title="Customer Segmentation & Product Recommendation",
    page_icon="🛒",
    layout="wide"
)

# LOAD SAVED MODELS

@st.cache_resource
def load_models():

    scaler = pickle.load(
        open("/content/CustomerSegmentation/scaler.pkl", "rb")
    )

    kmeans = pickle.load(
        open("/content/CustomerSegmentation/kmeans.pkl", "rb")
    )

    similarity_df = pickle.load(
        open("/content/CustomerSegmentation/product_similarity.pkl", "rb")
    )

    return scaler, kmeans, similarity_df


scaler, kmeans, similarity_df = load_models()

# SIDEBAR

st.sidebar.title("Navigation")

page = st.sidebar.radio(
    "Go To",
    [
        "🏠 Home",
        "🔍 Customer Segmentation",
        "🎯 Product Recommendation"
    ]
)

# HOME PAGE

if page == "🏠 Home":

    st.title(
        "🛒 E-Commerce Analytics System"
    )

    st.markdown("---")

    st.subheader(
        "Customer Segmentation & Product Recommendation"
    )

    st.write("""
    This application provides:

    ✅ Customer Segmentation using RFM Analysis and KMeans Clustering

    ✅ Product Recommendations using Item-Based Collaborative Filtering

    ---

    Modules Available:

    🔍 Customer Segmentation

    🎯 Product Recommendation
    """)

# CUSTOMER SEGMENTATION

elif page == "🔍 Customer Segmentation":

    st.title(
        "🔍 Customer Segmentation"
    )

    st.markdown(
        "Enter customer RFM values below."
    )

    recency = st.number_input(
        "Recency (days since last purchase)",
        min_value=0,
        value=30
    )

    frequency = st.number_input(
        "Frequency (number of purchases)",
        min_value=1,
        value=5
    )

    monetary = st.number_input(
        "Monetary (total spend)",
        min_value=0.0,
        value=1000.0
    )

    if st.button(
        "Predict Segment"
    ):

        customer_data = np.array([
            [
                recency,
                frequency,
                monetary
            ]
        ])

        scaled_data = scaler.transform(
            customer_data
        )

        cluster = kmeans.predict(
            scaled_data
        )[0]


       # UPDATE THESE AFTER CLUSTER PROFILING

        segment_map = {

            0: "Regular Customer",

            1: "At-Risk Customer",

            2: "Occasional Customer",

            3: "High-Value Customer"
        }

        segment = segment_map.get(
            cluster,
            "Unknown"
        )

        st.success(
            f"Customer Segment: {segment}"
        )

# PRODUCT RECOMMENDATION

elif page == "🎯 Product Recommendation":

    st.title(
        "🎯 Product Recommendation"
    )

    st.markdown(
        "Enter a product name to get recommendations."
    )

    product_name = st.text_input(
        "Product Name"
    )

    if st.button(
        "Get Recommendations"
    ):

        if product_name not in similarity_df.index:

            st.error(
                "Product not found."
            )

        else:

            recommendations = (

                similarity_df[product_name]

                .sort_values(
                    ascending=False
                )

                .iloc[1:6]

                .index

                .tolist()
            )

            st.subheader(
                "Recommended Products"
            )

            for product in recommendations:

                st.markdown(
                    f"""
                    <div style="
                        padding:10px;
                        margin:8px;
                        border-radius:10px;
                        background-color:#f2f2f2;
                    ">
                    ✅ {product}
                    </div>
                    """,
                    unsafe_allow_html=True
                )
```
